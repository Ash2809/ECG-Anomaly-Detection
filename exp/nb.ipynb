```python
import numpy as np
import pandas as pd
import wfdb
from wfdb import processing
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

record = wfdb.rdrecord('mit-bih-arrhythmia/100') 
annotation = wfdb.rdann('mit-bih-arrhythmia/100', 'atr')

ecg_signal = record.p_signal[:, 0] 
sampling_rate = record.fs  # Sampling frequency

scaler = StandardScaler()
ecg_signal = scaler.fit_transform(ecg_signal.reshape(-1, 1)).flatten()

plt.figure(figsize=(10, 4))
plt.plot(ecg_signal[:1000]) 
plt.title('ECG Signal Sample')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```
